```python
# Drop 'production_companies' column
movies_df = movies_df[['budget', 'genre', 'id', 'keywording', 'original_language', 'original_title', 'overview', 'popularity', 'prod_companies', 'production_countries', 'release_date', 'revenue', 'runtime', 'spoken_languages', 'status', 'vote_average', 'vote_count', 'cast', 'crew']]

# Display dataframe
movies_df.head(3)
```
