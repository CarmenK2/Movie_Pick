```python
# Import Dependencies 
import pandas as pd
```

```python
# Get Data
data_path = "Output/movies_cleaned2.csv"
df = pd.read_csv(data_path)
```

```python
# Get Survey Data
sd_df = pd.read_csv('Resources/Survey_Mood_MovieGenre.csv')
sd_df.head()
```

```python
# Function to Collate all the data based on column name 
def merge_profile(df, movie, genre):
    movie_col = "Movie_" + movie
    genre_col = "Movie_" + genre + '_genre'


    # pick the columns from the dataframe 
    df_merge = df[['ID','Mood','Age',movie_col,genre_col]]

    df_merge = df_merge.rename(columns = {movie_col:'title',genre_col:'genre'})

    return df_merge
```

```python
# Iterate through survey data and merge 

# Initiate empty list to store 
data_lst = []

# Iterate through 5 times as we know we have a static number of columns
for x in range(4):
    # Ensure list is clear, no residual data
    data_lst.clear()

    # Calling function and filling up list
    df_ = merge_profile(sd_df,str(x+1),str(x+1))
    data_lst.append(df_.values.tolist())

# re-instate dataframe 
merged_df = pd.DataFrame(data_lst[0],columns = ['id','mood','age','title','genre'])
```
