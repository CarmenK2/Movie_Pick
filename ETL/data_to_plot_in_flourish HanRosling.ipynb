```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read the csv file into a pandas DataFrame

data_df = pd.read_csv('Resources/movies_cleaned3.csv')
data_df.head(3)
```

```python
#check column names
data_df.columns
```

```python
#Drop out columns and choose "release_year", 'prod_companies','revenue'
data_df=data_df[['original_title','release_year','runtime','revenue','casting']]
data_df
```

```python
#Drop all Null value
data_df=data_df.dropna()
data_df
```

```python
#Fix up the index
data_df=data_df.reset_index()
data_df
```

```python
#Check the columns name
data_df.columns
```

```python
#Split the 'prod_companies' by commas and expand into new columns
data_1=data_df['casting'].str.split(',',n=6,expand=True)
data_1

```

```python
#Reset the index for merging later on
data_1=data_1.reset_index()
data_1
```

```python
#Rename the column 0 to prod_companies_x
data_1=data_1.rename(columns={0:"main_character"})

data_1

```

```python
#Merging two df together on "index column"
df_merge=pd.merge(data_1,data_df,on="index")
df_merge
```

```python
#Drop out necessary columns, keep colmns name of 'prod_companies_x','release_year','revenue'
df_merge1 = df_merge[['original_title','main_character','release_year','runtime','revenue']]
df_merge1
```

```python
df_merge1=df_merge1.reset_index()
df_merge1
```

```python
df_merge2=df_merge1['revenue'].div(1000000).round(2)
df_merge2
```

```python
df_merge3=pd.DataFrame({"Revenue_in_Million":df_merge2})
df_merge3
```

```python
df_merge3=df_merge3.reset_index()
df_merge3
```

```python
#Merging two df together on "index column"
df_final=pd.merge(df_merge3,df_merge1,on="index")
df_final
```

```python
df_final=df_final.reindex(columns=['original_title','release_year','main_character','runtime','Revenue_in_Million'])
df_final
```

```python
df_final.to_csv("Resources/data_readytoplot_MainC.csv")
```
