```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
survey_Mood = pd.read_csv('Resources/Survey_Mood_MovieGenre_Cleanned_Matched.csv')
survey_Mood.head()
```

```python
survey_Mood_df=survey_Mood
survey_Mood_df
```

```python
survey_Mood_df=survey_Mood_df[['Mood','Movie']]
survey_Mood_df
```

```python
survey_Mood_df=survey_Mood_df.rename(columns={'Movie':'original_title'})
survey_Mood_df.head()

```

```python
maindataset_df = pd.read_csv('Output/movies_cleaned2.csv')
maindataset_df.head()
```

```python

df_3 = pd.merge(survey_Mood_df, maindataset_df, on=['original_title'])
df_3.head()
```

```python
df_3=df_3[['Mood','original_title','id',"release_year","genre","overview","runtime","vote_average","budget","revenue",]]
df_3.head(20)
```

```python
df_3.to_csv("Output/combined_df.csv")
```
