```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read the csv file into a pandas DataFrame

data_df = pd.read_csv('../ML/results_with_mood.csv')
data_df.head(3)
```

```python
#Read in the mood_index file
mood_df=pd.read_csv('../Output/mood_index.csv')
mood_df.head(2)
```

```python
#Choosing the first item from column genre
def sel_first(x):
    return x.split(",")[0]
data_df['genre']=data_df.loc[:,"genre"].fillna("Adventure").apply(sel_first)
```

```python
#Convert runtime into nearest 0.5 ie: 1,1.5,2,2.5,3
data_df['runtime']=round((data_df["runtime"]/60 *2))/2
data_df
```

```python
#merging two dataframes using "prediction" from data_df and latch on "label" from the mood_df
data_df = pd.merge(data_df,mood_df, how='left', left_on="prediction", right_on="label")
data_df.head(5)
```

```python
#Drop unwanted columns
data_to_filter_df = data_df.iloc[:,[1,2,3,4,6,8]]
```
